## **This notebook was used to test the capturing of the pictures related to each series. Each series was parsed to capture the 'cover url' of each series for loading into it's own table in the datebase.**

```python
import json
import pandas as pd
import datetime as dt
from imdb import IMDb
from imdb import helpers
ia = IMDb()
```

```python
# Declare a variable for a list of captured series. 

rick_n_morty = '2861424'
ren_n_stimpy = '0101178'
beevis_n_butthead = '0105950'
aeon_flux = '0111873' # in case of re search copy Æon Flux
celeb_deathmatch = '0208614'
daria = '0118298'
south_park = '0121955'
fam_guy = '0182576'
american_dad = '0397306'
king_hill = '0118375'
space_ghost = '0108937'
futurama = '0149460'
aqua_thf = '0297494'
archer = '1486217'
boondocks = '0373732'
metalocalypse = '0839188'
robot_chick = '0437745'
squidbillies = '0457146'
super_jail = '1031283'
big_mouth = '6524350'
bobs_burg = '1561755'
bojack = '3398228'
mr_pickles = '2950342'
venture = '0417373'
simpsons = '0096697'
spawn = '0118475'
```

```python
# Declare list.
curr_titles = ([rick_n_morty,
                ren_n_stimpy,
                beevis_n_butthead,
                aeon_flux,
                celeb_deathmatch,
                daria,
                south_park,
                fam_guy,
                american_dad,
                king_hill,
                space_ghost,
                futurama,
                aqua_thf,
                archer,
                boondocks,
                metalocalypse,
                robot_chick,
                squidbillies,
                super_jail,
                big_mouth,
                bobs_burg,
                bojack,
                mr_pickles,
                venture,
                simpsons,
                spawn])
```

```python
# Loop through and grab a movie object for each series. 
series = []
for i in curr_titles:
    gap_stop = ia.get_movie(i)
    series.append(gap_stop)
        

```

```python
# Test the length for any missed shows. 
len(series)
```

```python
# Loop through and complete try's and excepts for each series.
# Capturing the synopsis of each show, the title, the id, and the cover url. 

synopsis = []
pic_url = []
title = []
id = []
for i in series:
    try:
        gap_syn = i['plot'][0]
        synopsis.append(gap_syn)
    except KeyError:
        synopsis.append('No Synopsis Found in IMDb Database')
    try:
        gap_pic = i['cover url']
        pic_url.append(gap_pic)
    except KeyError:
        pic_url.append(None)
    try:
        gap_title = i['title']
        title.append(gap_title)
    except KeyError:
        title.append(None)
    try:
        gap_id = i.getID()
        id.append(gap_id)
    except KeyError:
        id.append(None)
data = pd.DataFrame({ 'id': id,
                     'title': title,
                     'pic_url': pic_url,
                     'synopsis': synopsis
    
})
    
```
