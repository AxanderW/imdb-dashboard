```python
series_id = []
title = []
seasons = []
avg_rating = []
num_votes = []
```

```python
for i in series:
    try:
        l_series_id = i.getID()
    except KeyError:
        l_series = None
    try:
        l_title = i['title']
    except KeyError:
        l_title = None
    try:
        l_seasons = i['seasons']
    except KeyError:
        l_seasons = None
    try:
        l_avg_rating = i['rating']
    except KeyError:
        l_avg_rating = None
    try:
        l_votes = i['votes']
    except KeyError:
        l_votes = None
    
    series_id.append(l_series_id)
    title.append(l_title)
    seasons.append(l_seasons)
    avg_rating.append(l_avg_rating)
    num_votes.append(l_votes)
    
    
```
